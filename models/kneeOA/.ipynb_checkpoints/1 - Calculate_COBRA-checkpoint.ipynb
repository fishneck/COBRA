```python
import os
import numpy as np
import pandas as pd
import nibabel as nib
# from nifti import NiftiImage
```

```python
def get_COBRA(file_path):
    test_pred_impaired_image_load = nib.load(file_path).get_fdata()
    sub = file_path.split('/')[-1].split('_')[0]
    split = file_path.split('/')[-3].split('_predictions')[0]
    confidences = []
    probs = []
    cartilage_prob = []
    for k in range(5):# 5 segmentation classes
        v = test_pred_impaired_image_load.max(axis=3)[test_pred_impaired_image_load.argmax(axis=3)==k]
        confidences.append(v.mean()/6) # averaging over 6 planes, see mpunet instructions 
        if k==2 or k==4: #pixel-wise cartilage segmentation probabilities
            np.random.seed(312)
            prob_sample = np.random.choice(v,size=1000,replace=True)
            cartilage_prob.append(prob_sample)
    return (sub, split, confidences, cartilage_prob)#(subject id, test split, confidence for each class, sample pixel-wise probabilities )        
```

```python
dirs = ['/scratch/by2026/OAI/MultiPlanarUNet/my_project/test_cls_predictions/nii_files/',
      '/scratch/by2026/OAI/MultiPlanarUNet/my_project/val_cls_predictions/nii_files/',
      '/scratch/by2026/OAI/MultiPlanarUNet/my_project/train_cls_predictions/nii_files/'
     ]
```

```python
result = {}
result['sub'] = []
result['split'] = []
for k in range(5):
    result[f'confidence_{k}_mean'] = []
for k in [2,4]:
    result[f'cartilage_prob_{k}'] = []
for d in dirs:
    for f in os.listdir(d):
        sub, split, confidences, cartilage_prob = get_COBRA(os.path.join(d+f))
        result['sub'].append(sub)
        result['split'].append(split)
        for k in range(5):
            result[f'confidence_{k}_mean'].append(confidences[k])
            if k==2 or k==4:
                result[f'cartilage_prob_{k}'].append(cartilage_prob[k//2 - 1]) #2-->0 4-->1
```

```python
result_df = pd.DataFrame(result)
result_df.sort_values('sub')
```

```python
result_df.to_csv('knee_OA_confidence.csv',index=False)
```
